```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
#nltk.download("punkt")
def clean_text(text):
    if isinstance(text,str):
        
        text = re.sub(r"[^a-zA-Z0-9\s]","",text)
        
        #return word_tokenize(text.lower())
        return text.lower()
    
    return None

def get_compound_score(text):
    sia = SentimentIntensityAnalyzer()
    return sia.polarity_scores(text)["compound"]
    
def get_neg_score(text):
    sia = SentimentIntensityAnalyzer()
    return sia.polarity_scores(text)["neg"]

def get_neu_score(text):
    sia = SentimentIntensityAnalyzer()
    return sia.polarity_scores(text)["neu"]

def get_pos_score(text):
    sia = SentimentIntensityAnalyzer()
    return sia.polarity_scores(text)["pos"]

stocktweet_pd["cleaned_text"] = stocktweet_pd["tweet"].apply(clean_text)
stocktweet_pd=stocktweet_pd[['date','stock_name','tweet','cleaned_text']]
stocktweet_pd['compund']=stocktweet_pd['cleaned_text'].apply(lambda x: get_compound_score(x))
stocktweet_pd['neg']=stocktweet_pd['cleaned_text'].apply(lambda x: get_neg_score(x))
stocktweet_pd['neu']=stocktweet_pd['cleaned_text'].apply(lambda x: get_neu_score(x))
stocktweet_pd['pos']=stocktweet_pd['cleaned_text'].apply(lambda x: get_pos_score(x))
```

def agg_groups_day(stck_tweet):
    stck_tweet=stck_tweet.set_index('date')
    stck_tweet=stck_tweet.drop(columns='stock_name')
    stck_tweet=stck_tweet.drop(columns='tweet')
    stck_tweet=stck_tweet.groupby(pd.Grouper(freq='1D')).agg({'cleaned_text':'min','compund':'mean','neg':'mean','neu':'mean','pos':'mean'})
    stck_tweet=stck_tweet.ffill()
    return stck_tweet

```python
# Creating a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Creating a function to get Polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity



# Creating a new column 
df_combined['Subjectivity'] = df_combined['Ticker'].apply(getSubjectivity)
df_combined['Polarity'] = df_combined['Tweet'].apply(getPolarity)

# Displaying data
df_combined
```

# Group Polarity

```python
# Grouping the range of Polarity into different categories
def getInsight(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"

df_combined["Insight"] = df_combined["Sentiment Score"].apply(getInsight)
df_combined.head(50)
```

# Plot the Sentiment Scores vs Insights

```python
#Bar Plot - Top 5
plt.figure(figsize=(15, 8))
bar_plot = sns.barplot(x="Insight", y="Sentiment Score", hue="Ticker", data=df_combined, ci=None, palette="magma")

# Annotating each bar with the total rail line length
for p in bar_plot.patches:
    height = p.get_height()
    bar_plot.annotate(f"{height:.1f}",
                      (p.get_x() + p.get_width() / 2., height),
                      ha="center", va="center", xytext=(0, 9), textcoords="offset points")

plt.title(f"Insights vs Sentiment Scores")
plt.xlabel("Insight")
plt.ylabel("Sentiment Score")
plt.legend(title="Tickers", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.xticks(rotation=45, ha="right")  
plt.show()
```
