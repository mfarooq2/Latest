```python
import matplotlib.pyplot as plt
import pandas as pd
AAPL_tweets=pd.read_csv('AAPL_tweets.csv')
AAPL_stocks=pd.read_csv('AAPL_stocks.csv')
from textblob import TextBlob
import numpy as np
```

```python
AAPL_tweets
```

```python
AAPL_stocks
# Converting 'date' to datetime format
AAPL_stocks['Date'] = pd.to_datetime(AAPL_stocks['Date'])

# Sort by date to ensure time series are in order
AAPL_stocks.sort_values(by='Date', inplace=True)

# Setting date as index for easier time-based plotting
AAPL_stocks.set_index('Date', inplace=True)

AAPL_stocks=AAPL_stocks.groupby(pd.Grouper(freq='D')).mean()
AAPL_stocks=AAPL_stocks.ffill()
# Plotting Stock Price Trend
plt.figure(figsize=(12, 6))
plt.plot(AAPL_stocks.index, AAPL_stocks['Close'], label='Close Price', color='blue')
plt.title('AAPL Stock Price Trend')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Plotting Trading Volume Trend
plt.figure(figsize=(12, 6))
plt.plot(AAPL_stocks.index, AAPL_stocks['Volume'], label='Volume', color='orange')
plt.title('AAPL Trading Volume Trend')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import re
def clean_text(text):
    if isinstance(text,str):
        
        text = re.sub(r"[^a-zA-Z0-9\s]","",text)
        
        #return word_tokenize(text.lower())
        return text.lower()
    
    return None
```

```python
AAPL_tweets['date'] =pd.to_datetime(AAPL_tweets["date"],format='mixed',dayfirst=True)
AAPL_tweets["tweet"] = AAPL_tweets["tweet"].apply(clean_text)
AAPL_Sentiments=AAPL_tweets.copy()
AAPL_Sentiments['sentiments']= AAPL_Sentiments['tweet'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
# Sentiment analysis on each tweet for polarity and subjectivity
AAPL_Sentiments['polarity'] = AAPL_Sentiments['tweet'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
AAPL_Sentiments['subjectivity'] = AAPL_Sentiments['tweet'].apply(lambda x: TextBlob(str(x)).sentiment.subjectivity)

AAPL_Sentiments=AAPL_Sentiments[['date','sentiments','polarity','subjectivity']]

AAPL_Sentiments.sort_values(by='date', inplace=True)

# Setting date as index for easier time-based plotting
AAPL_Sentiments.set_index('date', inplace=True)
AAPL_Sentiments=AAPL_Sentiments.groupby(pd.Grouper(freq='D'))[['sentiments','polarity','subjectivity']].mean()
AAPL_Sentiments=AAPL_Sentiments.ffill()
```

```python
AAPL_Sentiments
```

```python


# Sentiment Analysis on Tweets
# Adding a 'sentiment' column based on polarity score from TextBlob

# Plotting Sentiment over Time
plt.figure(figsize=(12, 6))
plt.plot(AAPL_Sentiments.index, AAPL_Sentiments['sentiments'], label='Sentiment Polarity', color='purple')
plt.title('Tweet Sentiment Polarity Over Time')
plt.xlabel('Date')
plt.ylabel('Sentiment Polarity')
plt.legend()
plt.grid(True)
plt.show()

# Adding Moving Averages for Stock Price
AAPL_stocks['short_mavg'] = AAPL_stocks['Close'].rolling(window=20).mean()  # 20-day moving average
AAPL_stocks['long_mavg'] = AAPL_stocks['Close'].rolling(window=50).mean()   # 50-day moving average

# Plotting Stock Price with Moving Averages
plt.figure(figsize=(12, 6))
plt.plot(AAPL_stocks.index, AAPL_stocks['Close'], label='Close Price', color='blue')
plt.plot(AAPL_stocks.index, AAPL_stocks['short_mavg'], label='20-Day Moving Average', color='green')
plt.plot(AAPL_stocks.index, AAPL_stocks['long_mavg'], label='50-Day Moving Average', color='red')
plt.title('AAPL Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data=pd.merge(AAPL_Sentiments,AAPL_stocks,how='inner',left_index=True, right_index=True)
```

```python
import seaborn as sns

# Calculate correlation between close price, volume, and sentiment
correlation_matrix = data[['Close', 'Volume', 'sentiments']].corr()

# Plotting the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", square=True)
plt.title("Correlation Matrix for Close Price, Volume, and Sentiment")
plt.show()

# Identifying days with extreme sentiment
# Define thresholds for high and low sentiment
high_sentiment_threshold = 0.5
low_sentiment_threshold = -0.5

# Filtering days with high and low sentiment
high_sentiment_days = data[data['sentiments'] >= high_sentiment_threshold]
low_sentiment_days = data[data['sentiments'] <= low_sentiment_threshold]

# Plotting Close Price on High and Low Sentiment Days
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Close Price', color='blue', alpha=0.3)
plt.scatter(high_sentiment_days.index, high_sentiment_days['Close'], color='green', label='High Sentiment Days')
plt.scatter(low_sentiment_days.index, low_sentiment_days['Close'], color='red', label='Low Sentiment Days')
plt.title("Close Price on High and Low Sentiment Days")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Adding a column for quarters
data['quarter'] = data.index.quarter

# Analyzing volume by quarter to observe any trading pattern trends
quarterly_volume = data.groupby('quarter')['Volume'].mean()

# Visualizing average trading volume per quarter
plt.figure(figsize=(8, 6))
quarterly_volume.plot(kind='bar', color='teal')
plt.title("Average Trading Volume by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Average Volume")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Analyzing average closing price by quarter
quarterly_close = data.groupby('quarter')['Close'].mean()

# Visualizing average closing price per quarter
plt.figure(figsize=(8, 6))
quarterly_close.plot(kind='bar', color='coral')
plt.title("Average Closing Price by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Average Close Price (USD)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python


# Adding remaining features as before
# 1. Moving Averages (MA and EMA)
data['MA_5'] = data['Close'].rolling(window=5).mean()  # 5-day MA
data['MA_20'] = data['Close'].rolling(window=20).mean()  # 20-day MA
data['EMA_5'] = data['Close'].ewm(span=5, adjust=False).mean()  # 5-day EMA
data['EMA_20'] = data['Close'].ewm(span=20, adjust=False).mean()  # 20-day EMA

# 2. Price Change Percentage
data['Price_Change'] = data['Close'].pct_change() * 100  # Daily percent change

# 3. Volatility (20-day Standard Deviation of closing prices)
data['Volatility'] = data['Close'].rolling(window=20).std()

# 4. Relative Strength Index (RSI)
delta = data['Close'].diff(1)
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
data['RSI'] = 100 - (100 / (1 + rs))

# 5. On-Balance Volume (OBV)
data['OBV'] = (np.sign(data['Close'].diff()) * data['Volume']).fillna(0).cumsum()

# 6. Sentiment Moving Average (7-day)
data['Sentiment_MA_7'] = data['sentiments'].rolling(window=7).mean()

# Creating visualizations for these features
# Plotting Moving Averages (5 and 20-day)
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Close Price', color='blue', alpha=0.5)
plt.plot(data.index, data['MA_5'], label='5-Day MA', color='green')
plt.plot(data.index, data['MA_20'], label='20-Day MA', color='red')
plt.title("AAPL Close Price with 5-Day and 20-Day Moving Averages")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.legend()
plt.grid(True)
plt.show()

# Plotting Volatility
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Volatility'], label='20-Day Volatility', color='purple')
plt.title("AAPL 20-Day Price Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility (Standard Deviation)")
plt.legend()
plt.grid(True)
plt.show()

# Plotting Relative Strength Index (RSI)
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['RSI'], label='RSI (14-Day)', color='orange')
plt.axhline(70, color='red', linestyle='--', label='Overbought')
plt.axhline(30, color='green', linestyle='--', label='Oversold')
plt.title("AAPL Relative Strength Index (RSI)")
plt.xlabel("Date")
plt.ylabel("RSI")
plt.legend()
plt.grid(True)
plt.show()

# Plotting On-Balance Volume (OBV)
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['OBV'], label='On-Balance Volume', color='brown')
plt.title("AAPL On-Balance Volume (OBV)")
plt.xlabel("Date")
plt.ylabel("OBV")
plt.legend()
plt.grid(True)
plt.show()

# Plotting Sentiment Moving Average
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['sentiments'], label='Daily Sentiment', color='purple', alpha=0.3)
plt.plot(data.index, data['Sentiment_MA_7'], label='7-Day Sentiment Moving Average', color='blue')
plt.title("AAPL Tweet Sentiment with 7-Day Moving Average")
plt.xlabel("Date")
plt.ylabel("Sentiment Polarity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Selecting relevant features for correlation analysis with 'close' price
features_for_correlation = data[['Close', 'MA_5', 'MA_20', 'EMA_5', 'EMA_20', 'Price_Change', 'Volatility', 'RSI', 'OBV', 'sentiments', 'Sentiment_MA_7']]

# Calculating the correlation matrix
correlation_matrix = features_for_correlation.corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", square=True)
plt.title("Feature Correlation Matrix with Close Price")
plt.show()

# Extracting the correlation of each feature with 'close' price
correlation_with_close = correlation_matrix['Close'].sort_values(ascending=False)
correlation_with_close
```

```python
# Generating advanced features for stock prediction based on the suggestions

# MACD (Moving Average Convergence Divergence)
# Short-term EMA (12-day) and Long-term EMA (26-day) for MACD
data['EMA_12'] = data['Close'].ewm(span=12, adjust=False).mean()
data['EMA_26'] = data['Close'].ewm(span=26, adjust=False).mean()
data['MACD'] = data['EMA_12'] - data['EMA_26']
data['Signal_Line'] = data['MACD'].ewm(span=9, adjust=False).mean()  # Signal line (9-day EMA of MACD)

# Average True Range (ATR)
high_low = data['Close'] - data['Close'].shift(1)  # Calculate daily range
data['ATR'] = high_low.rolling(window=14).mean()  # 14-day ATR

# Bollinger Bands (20-day)
data['BB_Middle'] = data['MA_20']
data['BB_Upper'] = data['BB_Middle'] + 2 * data['Volatility']  # 2 standard deviations above
data['BB_Lower'] = data['BB_Middle'] - 2 * data['Volatility']  # 2 standard deviations below

# Directional Movement Index (DMI) and Average Directional Index (ADX)
# Calculating +DI and -DI components
data['plus_dm'] = np.where((data['Close'] - data['Close'].shift(1)) > 0, data['Close'] - data['Close'].shift(1), 0)
data['minus_dm'] = np.where((data['Close'].shift(1) - data['Close']) > 0, data['Close'].shift(1) - data['Close'], 0)

# Smoothed +DI and -DI over a 14-day period
data['plus_di'] = 100 * data['plus_dm'].rolling(window=14).sum() / data['ATR']
data['minus_di'] = 100 * data['minus_dm'].rolling(window=14).sum() / data['ATR']

# Average Directional Index (ADX)
data['dx'] = 100 * abs(data['plus_di'] - data['minus_di']) / (data['plus_di'] + data['minus_di'])
data['ADX'] = data['dx'].rolling(window=14).mean()

# Keltner Channels (20-day EMA with ATR)
data['KC_Middle'] = data['EMA_20']
data['KC_Upper'] = data['KC_Middle'] + 1.5 * data['ATR']
data['KC_Lower'] = data['KC_Middle'] - 1.5 * data['ATR']

# Stochastic Oscillator (14-day)
data['Lowest_Low'] = data['Close'].rolling(window=14).min()
data['Highest_High'] = data['Close'].rolling(window=14).max()
data['Stochastic_Oscillator'] = 100 * (data['Close'] - data['Lowest_Low']) / (data['Highest_High'] - data['Lowest_Low'])

# Visualizing MACD and Bollinger Bands
# MACD Visualization
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['MACD'], label='MACD', color='blue')
plt.plot(data.index, data['Signal_Line'], label='Signal Line', color='red')
plt.title("AAPL MACD with Signal Line")
plt.xlabel("Date")
plt.ylabel("MACD Value")
plt.legend()
plt.grid(True)
plt.show()

# Bollinger Bands Visualization
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Close Price', color='blue', alpha=0.3)
plt.plot(data.index, data['BB_Middle'], label='20-Day MA (Middle Band)', color='green')
plt.plot(data.index, data['BB_Upper'], label='Upper Bollinger Band', color='red', linestyle='--')
plt.plot(data.index, data['BB_Lower'], label='Lower Bollinger Band', color='purple', linestyle='--')
plt.fill_between(data.index, data['BB_Lower'], data['BB_Upper'], color='grey', alpha=0.1)
plt.title("AAPL Bollinger Bands")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()


# Visualizing ADX for trend strength analysis
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['ADX'], label='ADX', color='purple')
plt.axhline(20, color='grey', linestyle='--', label='Weak Trend')
plt.axhline(25, color='green', linestyle='--', label='Strong Trend')
plt.title("AAPL Average Directional Index (ADX)")
plt.xlabel("Date")
plt.ylabel("ADX Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculating advanced sentiment features



# Daily sentiment statistics
data['Daily_Sentiment_Mean'] = data['polarity'].resample('D').mean()
data['Daily_Sentiment_Volatility'] = data['polarity'].resample('D').std()

# Daily positive/negative tweet counts based on polarity thresholds
data['positive_tweet'] = data['polarity'] > 0
data['negative_tweet'] = data['polarity'] < 0
daily_positive_count = data['positive_tweet'].resample('D').sum()
daily_negative_count = data['negative_tweet'].resample('D').sum()

# Sentiment momentum - day-over-day change in daily sentiment mean
data['Sentiment_Momentum'] = data['Daily_Sentiment_Mean'].diff()

# Moving averages for sentiment (3-day and 7-day)
data['Sentiment_MA_3'] = data['Daily_Sentiment_Mean'].rolling(window=3).mean()
data['Sentiment_MA_7'] = data['Daily_Sentiment_Mean'].rolling(window=7).mean()

# Extreme sentiment flags
data['High_Sentiment_Flag'] = data['Daily_Sentiment_Mean'] > 0.5
data['Low_Sentiment_Flag'] = data['Daily_Sentiment_Mean'] < -0.5

# Visualizations for sentiment features
# Daily Sentiment Mean with Moving Averages
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Daily_Sentiment_Mean'], label='Daily Sentiment Mean', color='purple', alpha=0.4)
plt.plot(data.index, data['Sentiment_MA_3'], label='3-Day Sentiment MA', color='blue')
plt.plot(data.index, data['Sentiment_MA_7'], label='7-Day Sentiment MA', color='green')
plt.title("Daily Sentiment Mean with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Sentiment Polarity")
plt.legend()
plt.grid(True)
plt.show()

# Daily Positive and Negative Tweet Counts
plt.figure(figsize=(12, 6))
plt.plot(daily_positive_count.index, daily_positive_count, label='Daily Positive Tweet Count', color='green', alpha=0.7)
plt.plot(daily_negative_count.index, daily_negative_count, label='Daily Negative Tweet Count', color='red', alpha=0.7)
plt.title("Daily Positive and Negative Tweet Counts")
plt.xlabel("Date")
plt.ylabel("Tweet Count")
plt.legend()
plt.grid(True)
plt.show()

# Sentiment Momentum
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Sentiment_Momentum'], label='Sentiment Momentum', color='orange')
plt.title("Sentiment Momentum (Day-over-Day Change in Sentiment Mean)")
plt.xlabel("Date")
plt.ylabel("Sentiment Momentum")
plt.legend()
plt.grid(True)
plt.show()
```
